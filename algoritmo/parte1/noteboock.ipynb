```python
# Função para calcular a moda usando classes (regra de Sturges)
def get_moda(x_opt: np.ndarray):
    # Calcula o número de classes pela regra de Sturges
    n = int(1 + np.log2(len(x_opt)))
    
    # Calcula o intervalo das classes
    range_class = (max(x_opt) - min(x_opt)) / n
    print(f"Intervalo de classe: {range_class}")
    
    # Inicializando o dicionário de frequências das classes
    dict_hist = {}
    
    # Preenchendo o dicionário com as frequências por classe
    for i in range(n):
        limite_inferior = min(x_opt) + i * range_class
        limite_superior = limite_inferior + range_class
        # Filtra os valores que caem no intervalo da classe atual
        freq = len(x_opt[(x_opt >= limite_inferior) & (x_opt < limite_superior)])
        dict_hist[f"{limite_inferior:.5f}/{limite_superior:.5f}"] = freq
    
    # Determinando a moda (classe com a maior frequência)
    moda = max(dict_hist, key=dict_hist.get)
    
    return dict_hist, moda
```
